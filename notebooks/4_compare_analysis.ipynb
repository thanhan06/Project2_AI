```python
#So sánh 3 Dataset – Heart, Penguins, Wine Quality

import pandas as pd
import matplotlib.pyplot as plt

# Nhập thủ công thông tin tóm tắt từ quá trình huấn luyện (có thể cập nhật giá trị thực tế)
summary_data = [
    {
        "Tên Dataset": "Heart Disease",
        "Samples": 297,
        "Features": 13,
        "Classes": 2,
        "Max Accuracy": 0.8500  
    },
    {
        "Tên Dataset": "Penguins",
        "Samples": 344,
        "Features": 5,
        "Classes": 3,
        "Max Accuracy": 0.9552  
    },
    {
        "Tên Dataset": "Wine Quality",
        "Samples": 4898,
        "Features": 11,
        "Classes": 7,
        "Max Accuracy": 0.6184  
    }
]

# Tạo bảng so sánh
comparison_df = pd.DataFrame(summary_data)
print("\nBảng so sánh đặc trưng các dataset:")
display(comparison_df)

# Vẽ biểu đồ độ chính xác cao nhất
plt.figure(figsize=(8, 5))
plt.bar(comparison_df["Tên Dataset"], comparison_df["Max Accuracy"], color="skyblue")
plt.title("So sánh độ chính xác cao nhất giữa các dataset")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
