```python
from nbclient import NotebookClient
from nbformat import read, write
from pathlib import Path

notebooks = [
    "notebooks/1_heart_disease.ipynb",
    "notebooks/2_penguins.ipynb",
    "notebooks/3_custom_dataset.ipynb"
]

for nb_path in notebooks:
    print(f"🔄 Running {nb_path}")
    try:
        with open(nb_path, encoding='utf-8') as f:
            nb = read(f, as_version=4)
            client = NotebookClient(nb)
            client.execute()
        with open(nb_path, "w", encoding='utf-8') as f_out:
            write(nb, f_out)
        print(f"✅ {nb_path} executed successfully.")
    except Exception as e:
        print(f"❌ Error in {nb_path}: {e}")
        break
```
